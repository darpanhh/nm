```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def Bisection(F,a,b,p=4):
    tol=0.5*10**(-p)
    maxitr=int(2*np.ceil((np.log10(b-a)-np.log10(tol))/np.log10(2)))
    if F(a)*F(b)>0:
        print("Error occured")
        return None
    if abs(F(a))<tol:
        return round(a,p)
    if abs(F(b))<tol:
        return round(b,p)
    for _ in range(maxitr):
        c=(a+b)/2
        if abs(F(c))<tol:
            return round(c,p)
        if F(c)*F(a) < 0:
            b=c
        else:
            a=c
    else:
        print("Error could not converge to a valid solution")
        return None
```

```python
def PlotFx(F,x0,xn,xp=None,n=201,ylimit=None):
    X=np.linspace(x0,xn,n)
    Y=F(X)
    plt.figure(figsize=(9,2.5))
    plt.plot(X,Y)
    plt.grid()
    plt.axhline(0,color="green",linestyle="--",linewidth=1)
    plt.scatter(xp,0,color="red")
    if ylimit is not None:
        plt.ylim(ylimit)
    plt.show()
```

```python
f1=lambda x: x*np.sin(x)+np.cos(x)
root = Bisection(f1,2,3)
print(f"Root is:{root}")
PlotFx(f1,-8,8,root,201,[-7.5,7.5])# list or tuple
```
