```python
import numpy as np
import matplotlib.pyplot as plt
```

Lagrange's interpolation

```python
def Lagranges(X,Y,xp):
    xval=np.array(X)
    yval=np.array(Y)
    m,n=len(xval),len(yval)
    if m!=n:
        print("Mismatch between data")
        return None
    yp=0
    for i in range(n):
        prod=yval[i]
        for j in range(n):
            if i!=j:
                prod*=(xp-xval[j])/(xval[i]-xval[j])
        yp+=prod
    return yp
```

```python
xVals=[2,4,6,8,10]
yVals=[4,8,12,16,20]
xp=11
print(Lagranges(xVals,yVals,xp))
```

```python
plt.plot(xVals,yVals,marker='o',color='red')
X=np.linspace(min(xVals)-0.2,max(xVals)+0.2,5)
Y=Lagranges(xVals,yVals,X)
plt.plot(X,Y,marker='o',color='green')
```

Linear and Exponential Fit

```python
def LinearFit(xvalue,yvalue,n):
    X=np.array(xvalue,dtype=float)
    Y=np.array(yvalue,dtype=float)
    m,n=len(X),len(Y)
    if m!=n:
        print("Mismatch between data")
        return None
    sumx=np.sum(X)
    sumy=np.sum(Y)
    sumx2=np.sum(X**2)
    sumxy=np.sum(X*Y)
    Da=sumx2*sumy-sumx*sumxy
    Db=n*sumxy-sumx*sumy
    D=n*sumx2-(sumx)**2
    x=Da/D
    y=Db/D
    return x,y
    
```
